```python
import pandas as pd
import psycopg2
import plotly.express as px
```

<h1>Connect PostgreSQL

```python
conn = psycopg2.connect(
    host="localhost",
    database="stackoverflow_db",
    user="postgres",
    password="Codio590"
)

print("PostgreSQL Connected Successfully")
```

```python
query = """
SELECT 'Users' AS category, COUNT(*) AS total FROM users
UNION ALL
SELECT 'Questions', COUNT(*) FROM questions
UNION ALL
SELECT 'Answers', COUNT(*) FROM answers
UNION ALL
SELECT 'Comments', COUNT(*) FROM comments
UNION ALL
SELECT 'Tags', COUNT(*) FROM tags
"""

df = pd.read_sql(query, conn)

df
```

```python
fig = px.bar(
    df,
    x="category",
    y="total",
    title="Stack Overflow Dataset Overview"
)

fig.show()
```

```python
fig2 = px.pie(
    df,
    names="category",
    values="total",
    title="Stack Overflow Dataset Distribution"
)

fig2.show()
```
